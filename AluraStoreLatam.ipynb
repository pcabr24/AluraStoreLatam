```python
diccionario_tiendas = {
    "tienda": tienda,
    "tienda2": tienda2,
    "tienda3": tienda3,
    "tienda4": tienda4
}

print("--- 💸 Sumatoria Total de 'Precio' por Tienda ---")

# Diccionario para almacenar los resultados
suma_precios = {}

for nombre, df in diccionario_tiendas.items():
    try:
        # Calcular la suma de la columna 'Precio'
        total_precio = df['Precio'].sum()

        # Almacenar y mostrar el resultado
        suma_precios[nombre] = total_precio
        print(f"La sumatoria de 'Precio' en {nombre} es: {total_precio:,.2f}")

    except KeyError:
        # Manejo de error si alguna tienda no tiene la columna 'Precio'
        print(f"ADVERTENCIA: El DataFrame '{nombre}' no tiene la columna 'Precio'.")
    except TypeError:
        # Manejo de error si la columna 'Precio' tiene tipos de datos mixtos/incompatibles
        print(f"ERROR: La columna 'Precio' en '{nombre}' contiene datos incompatibles para la suma (revisar dtypes).")

print("-------------------------------------------------------------------")
```

# 3. Calificación promedio de la tienda


```python
# Suponiendo que tus DataFrames están en este diccionario:
diccionario_tiendas = {
    "tienda": tienda,
    "tienda2": tienda2,
    "tienda3": tienda3,
    "tienda4": tienda4
}

print("--- 🥇 Conteo Agrupado Ordenado por Mayor Frecuencia (Columna 'Unidades') ---")

for nombre, df in diccionario_tiendas.items():
    try:
        # 1. Agrupamos y contamos (usando .size()).
        conteo_serie = df.groupby(['Producto', 'Calificación']).size()

        # 2. CONVERTIMOS LA SERIE A DATAFRAME y ASIGNAMOS EL NOMBRE DE COLUMNA 'Unidades'.
        conteo_df = conteo_serie.to_frame(name='Unidades')

        # 3. Ordenamos de forma descendente por la nueva columna 'Unidades'.
        conteo_ordenado = conteo_df.sort_values(by='Unidades', ascending=False)

        print(f"\n📢 **Tienda: {nombre}**")
        print(conteo_ordenado)
        print("------------------------------------------")

    except KeyError as e:
        columna_faltante = str(e).strip("'")
        print(f"ADVERTENCIA: El DataFrame '{nombre}' no tiene la columna '{columna_faltante}' (o 'Producto').")
    except Exception as e:
        print(f"ERROR al procesar '{nombre}': {e}")
```

# 5. Envío promedio por tienda

```python
import pandas as pd

def analizar_promedio_envio(diccionario_tiendas):
    """
    Calcula y muestra el costo de envío promedio para cada tienda.

    Args:
        diccionario_tiendas (dict): Un diccionario de DataFrames (tiendas).

    Returns:
        dict: Un diccionario con los costos de envío promedio por tienda.
    """
    # --- Asumimos el nombre de columna correcto por tu ejecución anterior ---
    COLUMNA_ENVIO = 'Costo de envío'

    promedios_envio = {}

    print("--- 💰 Costo de Envío Promedio por Tienda ---")

    for nombre, df in diccionario_tiendas.items():
        print(f"\n📢 **Analizando: {nombre}**")
        try:
            # 1. Seleccionar la columna de costo de envío y aplicar .mean()
            costo_promedio = df[COLUMNA_ENVIO].mean()

            # Guardar el resultado
            promedios_envio[nombre] = costo_promedio

            print(f"   - Costo de Envío Promedio: {costo_promedio:,.2f}")

        except KeyError:
            print(f"   - ADVERTENCIA: El DataFrame '{nombre}' no tiene la columna '{COLUMNA_ENVIO}'.")
        except Exception as e:
            print(f"   - ERROR inesperado al procesar '{nombre}': {e}")

    print("---------------------------------------------------------")
    return promedios_envio

# --- EJECUCIÓN Y ALMACENAMIENTO DE RESULTADOS ---

# 1. Definir los datos de entrada
diccionario_tiendas = {
    "tienda": tienda,
    "tienda2": tienda2,
    "tienda3": tienda3,
    "tienda4": tienda4
}

# 2. Llamar a la función y guardar el resultado en la variable 'resultados_envio'
resultados_envio = analizar_promedio_envio(diccionario_tiendas)

print("\nResultados guardados en la variable 'resultados_envio':")
print(resultados_envio)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Asumimos que la columna de ingreso es 'Ventas'
COLUMNA_INGRESO = 'Precio'

diccionario_tiendas = {
    "tienda": tienda,
    "tienda2": tienda2,
    "tienda3": tienda3,
    "tienda4": tienda4
}

# Calcular el ingreso total de ventas para cada tienda
ingresos_totales = {}
for nombre, df in diccionario_tiendas.items():
    try:
        # Sumamos la columna de ingreso de ventas
        ingresos_totales[nombre] = df[COLUMNA_INGRESO].sum()
    except KeyError:
        print(f"ADVERTENCIA: La tienda '{nombre}' no tiene la columna '{COLUMNA_INGRESO}'.")
        ingresos_totales[nombre] = 0 # Asignar 0 si no existe para no romper el gráfico

# Convertir el diccionario a una Serie de Pandas
serie_ingresos = pd.Series(ingresos_totales)
```
